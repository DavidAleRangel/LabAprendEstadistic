## Advertaising.csv

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
 
# ── Carga de datos ──────────────────────────────────────────────────────────
df = pd.read_csv("Advertising.csv", index_col=0)
 
X = df[["TV", "radio", "newspaper"]]   # features
y = df["sales"]                         # target
```

```python
# Regresión Lineal Simple
lr_full = LinearRegression()
lr_full.fit(X, y)
y_pred_lr_full = lr_full.predict(X)
 
# R² de la Regresión Lineal 
r2_lr_full = r2_score(y, y_pred_lr_full)
print(f"\n[1] Regresión Lineal")
print(f"    Intercepto : {lr_full.intercept_:.4f}")
print(f"    Coeficientes: TV={lr_full.coef_[0]:.4f}, "
      f"radio={lr_full.coef_[1]:.4f}, newspaper={lr_full.coef_[2]:.4f}")
print(f"\n[2] R² Regresión Lineal  : {r2_lr_full:.4f}")
 
# Árbol de Decisión para Regresión
dt_full = DecisionTreeRegressor(random_state=42)
dt_full.fit(X, y)
y_pred_dt_full = dt_full.predict(X)
 
# R² del Árbol de Decisión 
r2_dt_full = r2_score(y, y_pred_dt_full)
print(f"\n[3] Árbol de Decisión (DecisionTreeRegressor)")
print(f"    Profundidad máxima alcanzada: {dt_full.get_depth()}")
print(f"\n[4] R² Árbol de Decisión : {r2_dt_full:.4f}")
```

```python
print("  PARTE 2 — Train/Test split 50/50")
 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=42
)
print(f"\n  Muestras entrenamiento : {len(X_train)}")
print(f"  Muestras prueba        : {len(X_test)}")
 
# Regresión Lineal (split)
lr_split = LinearRegression()
lr_split.fit(X_train, y_train)
y_pred_lr_train = lr_split.predict(X_train)
y_pred_lr_test  = lr_split.predict(X_test)
 
# R² de la Regresión Lineal (split) 
r2_lr_train = r2_score(y_train, y_pred_lr_train)
r2_lr_test  = r2_score(y_test,  y_pred_lr_test)
 
print(f"\n[1] Regresión Lineal (50/50)")
print(f"    Intercepto : {lr_split.intercept_:.4f}")
print(f"    Coeficientes: TV={lr_split.coef_[0]:.4f}, "
      f"radio={lr_split.coef_[1]:.4f}, newspaper={lr_split.coef_[2]:.4f}")
print(f"\n[2] R² Regresión Lineal")
print(f"    Train : {r2_lr_train:.4f}")
print(f"    Test  : {r2_lr_test:.4f}")
```

### Árbol de decisión

```python
dt_split = DecisionTreeRegressor(random_state=42)
dt_split.fit(X_train, y_train)
y_pred_dt_train = dt_split.predict(X_train)
y_pred_dt_test  = dt_split.predict(X_test)
```

### R² del Árbol de Decisión

```python
r2_dt_train = r2_score(y_train, y_pred_dt_train)
r2_dt_test  = r2_score(y_test,  y_pred_dt_test)
 
print(f"\n[3] Árbol de Decisión (50/50)")
print(f"    Profundidad máxima alcanzada: {dt_split.get_depth()}")
print(f"\n[4] R² Árbol de Decisión")
print(f"    Train : {r2_dt_train:.4f}")
print(f"    Test  : {r2_dt_test:.4f}")
```
