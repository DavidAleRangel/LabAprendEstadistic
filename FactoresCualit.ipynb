```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Agf = pd.read_excel('Aguasfeo.xlsx')
Agf
```

```python
Agf.columns
```

```python
Agf['Tamaño_bin'] = Agf['Tamaño'].map({'grande': 1, 'chica': 0})
Agf['Hielo_bin'] = Agf['Hielo'].map({'con': 1, 'sin': 0})
Agf['Kahlúa'] = Agf['Sabor '].map({'kahlúa': 1, 'horchata': 0})
Agf['Horchata'] = Agf['Sabor '].map({'horchata': 1, 'kahlúa':0})
Agf
```

```python
import statsmodels.api as sm
X = Agf.drop('Score',axis = 1)
X = X.drop('Tamaño',axis = 1)
X = X.drop('Hielo',axis = 1)
X = X.drop('Sabor ',axis = 1)

y = Agf['Score']

X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
Agf['Tamaño_bin'] = Agf['Tamaño'].map({'grande': -1, 'chica': 1})
Agf['Hielo_bin'] = Agf['Hielo'].map({'con': 1, 'sin': -1})
Agf['Kahlúa'] = Agf['Sabor '].map({'kahlúa': 1, 'horchata': 0})
Agf['Horchata'] = Agf['Sabor '].map({'horchata': 1, 'kahlúa':0})
Agf
```

```python
import statsmodels.api as sm
X = Agf.drop('Score',axis = 1)
X = X.drop('Tamaño',axis = 1)
X = X.drop('Hielo',axis = 1)
X = X.drop('Sabor ',axis = 1)

y = Agf['Score']

X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

### Parte 3 (Comparación Betas y observaciones)

Betas en la regresion con **numeros buleanos**:
*  $\beta_0$ = 4.3
*  $\beta_1$ = 2.45
*  $\beta_2$ = 0.55
*  $\beta_3$ = 2.125
*  $\beta_4$ = 2.175

Betas en la regresion con **numeros -1, 1**:
*  $\beta_0$ = 5.3
*  $\beta_1$ = -1.225
*  $\beta_2$ = 0.275
*  $\beta_3$ = 2.625
*  $\beta_4$ = 2.675

**Observaciones:**

El cambio en las betas ocurre principalmente por la escala y el punto de referencia de las variables. Al usar números booleanos (0 y 1), la diferencia entre categorías es de una unidad, por lo que la beta representa el cambio total al pasar de una opción a otra; en cambio, al usar −1 y 1, la distancia se duplica a dos unidades, lo que provoca que las betas de las pendientes se reduzcan a la mitad para compensar ese "paso" más largo. Además, el intercepto B0 se desplaza porque en el modelo booleano representa el valor base cuando todo es cero, mientras que en el modelo −1/1 se mueve hacia el promedio global de los datos. Finalmente, el cambio de signo en β1(de positivo a negativo) se debe por invertir la lógica de la variable: antes "grande" sumaba al ser el valor positivo (1) y ahora "chica" es la que tiene el valor positivo, invirtiendo el sentido de la relación.
