1.- Dos equipos juegan una serie de juegos (4 de 7, el que gane 4 juegos gana). Cada equipo tiene un 50% de probabilidad de ganar cada juego (no empates). ¿Cuál es la probabilidad de que la serie se vaya a 7 juegos? (5/16)

```python
# EJERCICIO 1

print("\n Ejercicio 1: Serie 4 de 7")

va_a_7 = 0
for _ in range(N):
    gana_a = gana_b = 0
    juegos = 0
    while gana_a < 4 and gana_b < 4:
        juegos += 1
        if random.random() < 0.5:
            gana_a += 1
        else:
            gana_b += 1
    if juegos == 7:
        va_a_7 += 1

prob_7 = va_a_7 / N
print(f"  Probabilidad de llegar a 7 juegos: {prob_7:.4f}")
print(f"  Valor teórico: {5/16:.4f}  (5/16)")
```

 2.- Una en mil personas tienen una enfermedad H, y la prueba de la enfermedad sale correcta 98% de las veces cuando la persona está enferma. Por otro lado, si la persona está enferma la prueba tiene un error del 1%. Si alguien se hace una prueba y sale positiva, ¿cuál es la probabilidad de que tengan la enfermedad? (98/1097)

```python
# EJERCICIO 2: Teorema de Bayes - Enfermedad H

print("\n Ejercicio 2: Prueba de enfermedad H")

# P(enferma) = 1/1000
# P(positivo | enferma)   = 0.98
# P(positivo | no enferma) = 0.01  ← "error del 1% si NO está enferma"
positivo_y_enfermo = 0
positivo_total = 0

for _ in range(N):
    enfermo = random.random() < 1 / 1000
    if enfermo:
        test_positivo = random.random() < 0.98
    else:
        test_positivo = random.random() < 0.01
    if test_positivo:
        positivo_total += 1
        if enfermo:
            positivo_y_enfermo += 1

prob_enfermo_dado_positivo = positivo_y_enfermo / positivo_total if positivo_total > 0 else 0
print(f"  P(enferma | positivo): {prob_enfermo_dado_positivo:.4f}")
print(f"  Valor teórico: {98/1097:.4f}  (98/1097)")
```

3.- Tres amigos diferentes te dijeron que está lloviendo, y cada persona tiene 1/3 de probabilidad de mentir. Si el pronóstico del clima dijo que la probabilidad de lluvia era 25%, ¿cuál es la probabilidad de que sí esté lloviendo? (8/11)

```python
#EJERCICIO 3

print("\n Ejercicio 3: Tres amigos y la lluvia")

# P(lluvia) = 0.25
# P(mentira) = 1/3  →  si llueve, cada amigo dice "llueve" con prob 2/3
#                       si NO llueve, cada amigo dice "llueve" con prob 1/3
lloviendo_y_todos_dicen_llueve = 0
todos_dicen_llueve = 0

for _ in range(N):
    lloviendo = random.random() < 0.25
    if lloviendo:
        # los tres dicen la verdad (llueve) si NO mienten
        todos_dicen = all(random.random() > 1/3 for _ in range(3))
    else:
        # los tres mienten (dicen llueve cuando no llueve)
        todos_dicen = all(random.random() < 1/3 for _ in range(3))
    if todos_dicen:
        todos_dicen_llueve += 1
        if lloviendo:
            lloviendo_y_todos_dicen_llueve += 1

prob_llueva = lloviendo_y_todos_dicen_llueve / todos_dicen_llueve if todos_dicen_llueve > 0 else 0
print(f"  P(lluvia | los 3 dicen lluvia): {prob_llueva:.4f}")
print(f"  Valor teórico: {8/11:.4f}  (8/11)")
```

5.- Tienes diez dados justos. Si arrojas todos al mismo tiempo, ¿cuál es la probabilidad de que la suma de los dados sea múltiplo de 6?

```python
# EJERCICIO 5

print("\n Ejercicio 5: 10 dados — suma múltiplo de 6")

multiple_de_6 = 0
for _ in range(N):
    total = sum(random.randint(1, 6) for _ in range(10))
    if total % 6 == 0:
        multiple_de_6 += 1

prob_mult6 = multiple_de_6 / N
print(f"  Probabilidad de suma múltiplo de 6: {prob_mult6:.4f}")
# Valor teórico ≈ 1/6 (uniforme para dados con muchas caras, aprox)
print(f"  Referencia 1/6 ≈ {1/6:.4f}")
```
