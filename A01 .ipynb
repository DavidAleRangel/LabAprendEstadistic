# Salarios

## Realiza regresiones polinomiales de orden 1, 3, 5 y 7.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
salarios = pd.read_csv('Salarios_minimos.csv')
salarios.head()
```

```python
x1sl = salarios["Periodo"]
x1sl = np.array(x1sl).reshape(-1, 1)
```

```python
y1sl = salarios["Salario"]
y1sl = np.array(y1sl).reshape(-1,1)
```

### Regresión Grado 1

```python
lr = LinearRegression()
X1 = np.hstack([x1sl])
lr.fit(X1,y1sl)
lr.intercept_
```

```python
lr.coef_
```

Para la regresion de **grado 1**:
*  $\beta_0$ = -5022.60
*  $\beta_1$ = 2.528

### Regresión Grado 3

```python
x2sl = x1sl * x1sl
x3sl = x2sl * x1sl
```

```python
X3 = np.hstack([x1sl,x2sl,x3sl])
```

```python
lr = LinearRegression()
lr.fit(X3,y1sl)
lr.intercept_
```

```python
lr.coef_
```

Para la regresion de **grado 3**:
*  $\beta_0$ = -5217722.95
*  $\beta_1$ = 7.83146
*  $\beta_2$ = -3.91934

```python
lr = LinearRegression()
lr.fit(X5,y1sl)
lr.intercept_
```

```python
lr.coef_
```

## Compara el R2 que te da cada modelo.

```python
from sklearn.metrics import r2_score

# Ajuste y R2 de cada modelo

# Modelo X1 (lineal)
lr1 = LinearRegression()
lr1.fit(X1, y1sl)
y_hat1 = lr1.predict(X1)
r2_X1 = r2_score(y1sl, y_hat1)

# Modelo X3 (grado 3)
lr3 = LinearRegression()
lr3.fit(X3, y1sl)
y_hat3 = lr3.predict(X3)
r2_X3 = r2_score(y1sl, y_hat3)

# Modelo X5 (grado 5)
lr5 = LinearRegression()
lr5.fit(X5, y1sl)
y_hat5 = lr5.predict(X5)
r2_X5 = r2_score(y1sl, y_hat5)

# Modelo X7 (grado 7)
lr7 = LinearRegression()
lr7.fit(X7, y1sl)
y_hat7 = lr7.predict(X7)
r2_X7 = r2_score(y1sl, y_hat7)

# Mostrar resultados
print("R² de cada modelo:")
print(f"Modelo X1 (Grado1):   {r2_X1:.4f}")
print(f"Modelo X3 (Grado 3):  {r2_X3:.4f}")
print(f"Modelo X5 (Grado 5):  {r2_X5:.4f}")
print(f"Modelo X7 (Grado 7):  {r2_X7:.4f}")
```

## Crea una gráfica con una nube de dispersión (datos originales) y tus modelos.

```python
# Ajuste de modelos
# Modelo lineal (X1)
lr1 = LinearRegression()
lr1.fit(X1, y1sl)
y_hat1 = lr1.predict(X1)

# Modelo cúbico (X3)
lr3 = LinearRegression()
lr3.fit(X3, y1sl)
y_hat3 = lr3.predict(X3)

# Modelo grado 5 (X5)
lr5 = LinearRegression()
lr5.fit(X5, y1sl)
y_hat5 = lr5.predict(X5)

# Modelo grado 7 (X7)
lr7 = LinearRegression()
lr7.fit(X7, y1sl)
y_hat7 = lr7.predict(X7)

# Gráfica

plt.figure(figsize=(10,6))

# Nube de dispersión (datos originales)
plt.scatter(x1sl, y1sl, label="Datos originales", alpha=0.7)

# Curvas estimadas
plt.plot(x1sl, y_hat1, label="Modelo X1 (Grado 1)", linewidth=2)
plt.plot(x1sl, y_hat3, label="Modelo X3 (Grado 3)", linewidth=2)
plt.plot(x1sl, y_hat5, label="Modelo X5 (Grado 5)", linewidth=2)
plt.plot(x1sl, y_hat7, label="Modelo X7 (Grado 7)", linewidth=2)

plt.xlabel("Periodo")
plt.ylabel("Salario mínimo")
plt.title("Salario mínimo: Datos originales vs Modelos Polinomiales")
plt.legend()
plt.grid(True)

plt.show()
```
